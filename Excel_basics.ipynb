#Excel basics
1) What is differance between the Relative, Absoulte, and Mixed cell references?

- **Relative cell referances**- referances get change automatically as the formula get copy
- e.g. for calculation of the average, total,Etc.
- **Absolute cell referancing**- we fix the one cell in the sheet with the symbol of $
- e.g for calculations of the percentages.
- **Mixed cell referencing**- one part is fixed and other changes
